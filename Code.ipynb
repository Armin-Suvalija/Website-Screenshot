```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
import time
import os
import re

websites = [   
   "https://aithor.com/",
    "https://undetectable.ai/",
    "https://spinbot.com",
    "https://csgenerator.com",
    "https://dailywritingtips.com",
    "https://essaytyper.com",
    "https://quickanddirtytips.com/grammar-girl-archive",
    "https://grammarbook.com",
    "https://grammarly.com/blog",
    "https://writingcenter.fas.harvard.edu/",
    "https://hemingwayapp.com",
    "https://hemingwayapp.com",
    "https://www.instaprompt.ai/?ref=producthunt",
    "https://iwl.me",
    "https://noredink.com",
    "https://www.phrasebank.manchester.ac.uk/",
    "https://polishmywriting.com",
    "https://prowritingaid.com",
    "https://quetext.com",
    "https://readable.com",
    "https://www.slickwrite.com",
    "https://topessaywriting.com",
    "https://turnitin.com",
    "https://chromewebstore.google.com/detail/wordtune-ai-writing-parap/nllcnknpjnininklegdoijpljgdjkijc",
    "https://transformer.huggingface.co",
    "https://www.writeboxapp.com",
    "https://writecheck.com",
    "https://writeroomapp.com",
    "https://zenpen.io/",
    "https://donthitpublish.growthtools.com/",
    "https://chrome.google.com/webstore/detail/after-the-deadline/fcdjadjbdihbaodagojiomdljhjhjfho",
    "https://bear.app/"
]

save_path = "Screenshots/WRITING"
os.makedirs(save_path, exist_ok=True)

width = 1382
height = 915

chrome_options = Options()
chrome_options.add_argument(f"--window-size={width},{height}")
chrome_options.add_argument("--headless")
chrome_options.add_argument("--force-device-scale-factor=1")  # Force scale factor
chrome_options.add_argument("--headless=new")  # Use new headless mode

driver_path = "chromedriver-win64/chromedriver.exe"

# Initialize WebDriver
service = Service(driver_path)
driver = webdriver.Chrome(service=service, options=chrome_options)

# Ensure the window size is set explicitly
driver.set_window_size(width, height)

def extract_domain_name(url):
    """ Extracts the domain name from a URL and formats it for use as a filename. """
    # Remove "https://", "www.", and split by the first "/"
    domain = re.sub(r"^(https?://)?(www\.)?", "", url)
    domain = domain.split("/")[0]  # Get the domain part
    domain = domain.lower()  # Convert to lowercase
    return domain

# Loop through each website and take a screenshot
for index, url in enumerate(websites):
    try:
        driver.get(url)
        time.sleep(3)  # Wait for the page to load
        
        # Extract the domain name and sanitize it
        domain_name = extract_domain_name(url)
        
        # Create the screenshot filename
        screenshot_filename = os.path.join(save_path, f"{domain_name}.png")
        
        # Save the screenshot
        driver.save_screenshot(screenshot_filename)
        print(f"Saved screenshot: {screenshot_filename}")
    except Exception as e:
        print(f"Error capturing {url}: {e}")

driver.quit()
```
